```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import cv2
from matplotlib import pyplot as plt
from joblib import Parallel, delayed
import glob
import random
import albumentations as A

IMG_SIZE = 256
filenames = glob.glob( f'../input/jpeg-melanoma-{IMG_SIZE}x{IMG_SIZE}-without-hair/train/*' )
```

```python
transform = A.Compose([
    A.OneOf([
        A.Flip(p=0.5),
        A.IAAFliplr(p=0.5),
        A.Transpose(p=0.5),
        A.IAAFlipud(p=0.5),
    ], p=0.5),
    A.OneOf([
        A.Rotate(limit=365, p=0.5),
        A.ShiftScaleRotate(p=0.5),
    ], p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.05, contrast_limit=0.05, p=0.5),
    A.Blur(blur_limit=3, p=0.5),
    A.GaussNoise(p=0.5),
    A.Cutout(num_holes=8, max_h_size=int(IMG_SIZE * 0.125), max_w_size=int(IMG_SIZE * 0.125), p=0.5),
])
```

```python
plt.figure(figsize=(20,8))
for i,k in enumerate(random.sample(filenames, 5)):
    img = cv2.imread( k )
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(2,5,i+1); plt.axis('off')
    plt.imshow(img)
    augmented_image = transform(image=img)['image']
    plt.subplot(2,5,i+6); plt.axis('off')
    plt.imshow(augmented_image)
    
plt.show()
```
