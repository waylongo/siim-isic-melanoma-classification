```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score
import cv2

IMAGE_SIZE = 384
oof_file_name = "../ensembles/oofs/oof_imgsize_384_eff_b4.csv"
df_oof = pd.read_csv(oof_file_name)

print(df_oof.shape)
df_oof.head()
```

```python
df_oof["oof_rank"] = df_oof.oof.rank(pct=True)

print(df_oof.shape)
df_oof.head()
```

```python
oof_auc = roc_auc_score(df_oof.target, df_oof.oof)
oof_rank_auc = roc_auc_score(df_oof.target, df_oof.oof_rank)
print(f"oof orig score is {oof_auc:.3f}")
print(f"oof rank score is {oof_rank_auc:.3f}")
```

```python
RANGE = [0.0, 0.7]

df_mistake = df_oof[(df_oof.target == 1) & (df_oof.oof_rank > RANGE[0]) & (df_oof.oof_rank < RANGE[1])]

print(f"mistake length is {len(df_mistake)}")

df_mistake_sample = df_mistake.sample(10)
mistake_name_list = df_mistake_sample.image_name.values
mistake_prob = df_mistake_sample.oof_rank.values
```

```python
plt.figure(figsize=(20,8))
for i,k in enumerate(mistake_name_list):
    img = cv2.imread(f"../input/jpeg-melanoma-{IMAGE_SIZE}x{IMAGE_SIZE}-without-hair/train/{k}.jpg")
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    plt.subplot(2,5,i+1); plt.axis('off')
    plt.imshow(img)
    plt.title(f"{k} - {mistake_prob[i]:.2f}")
plt.show()
```
